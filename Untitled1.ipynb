```python
sns.pairplot(df, x_vars = ['displacement', 'horsepower', 'weight', 'acceleration', 'mpg'], y_vars = 'mpg')
```

```python
y = df['mpg']
```

```python
y.shape
```

```python
x = df[['displacement', 'horsepower', 'weight', 'acceleration']]
```

```python
x.shape
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x
```

```python
pd.DataFrame(x).describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size = 0.7)
```

```python
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
#Linear Regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#train or fit model
lr.fit(xtrain, ytrain)
```

```python
lr.intercept_
```

```python
lr.coef_
```

Mileage = 23.6 - 1.14Displacement - 1.06Horsepower - 4.63Weight -0.39Accelration + error

prediction

```python
y_pred = lr.predict(xtest)
```

```python
y_pred
```

Model accuracy

```python
from sklearn.metrics import mean_absolute_error, r2_score, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(ytest, y_pred)
```

```python
mean_absolute_percentage_error(ytest, y_pred)
```

```python
r2_score(ytest, y_pred)
```

Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 2, interaction_only = True, include_bias = False)
```

```python
x_train2 = poly.fit_transform(xtrain)
```

```python
x_test2 = poly.fit_transform(xtest)
```

```python
lr.fit(x_train2, ytrain)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred_poly = lr.predict(x_test2)
```

Model accuracy

```python
mean_absolute_error(ytest, y_pred_poly)
```

```python
mean_absolute_percentage_error(ytest, y_pred_poly)
```

```python
r2_score(ytest, y_pred_poly)
```
